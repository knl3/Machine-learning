```python
"""
Q 2
"""

insurance = pd.read_csv('InsurancePurchase.csv', sep=',', header=0)
insurance['intercept']=1
insurance['Purchase[Yes]']=insurance['Purchase'].astype('category').cat.codes
insurance['Purchase[No]']=insurance['Purchase'].astype('category').cat.codes

y=insurance['Purchase[No]']
X = insurance[['MOPLHOOG','MBERHOOG','MINK7512','MINK123M','intercept']]


# train test split
# to avoid over fitting, we need to divide dataset into training and 
# test splits so that it can give us a better idea as to how our algorithm
# is tested on un-seen data
# testing the dataset for first 1001 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1000, random_state=60)


# before making actual predictions, good to scale features so that all of 
# them are uniformly evaluated
scaler = StandardScaler()  
scaler.fit(X_train)
X_train = scaler.transform(X_train)  
X_test = scaler.transform(X_test) 

## knn classification
classifier = KNeighborsClassifier(n_neighbors=5)  
classifier.fit(X_train, y_train)  

## predict test set 
y_pred = classifier.predict(X_test) 

# evaluate the algorithm
print(metrics.confusion_matrix(y_test, y_pred)) 
print("Accuracy:",classifier.score(X_test,y_test))

from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))


```

```python
"""
Q 3-1. Try to fit a logistic regression model and evaluate its performance
on the test data. What is the prediction accuracy when you use 0.5 as the 
predicted probability cut-off for the classifier? What is prediction 
accuracy among the customers that are predicted to buy insurance?

Chosen variables that may affect purchase of insurance:
MOPLHOOG: High level education
MBERHOOG: High status
MINK7512: Income 75-122.000
MINK123M: > 123.000

Q 3-2. Try other values (e.g. 0.1,0.2,0.3â€¦) as the probability threshold 
for the classifier.  What is prediction accuracy among the customers that 
are predicted to buy insurance?
"""


insurance = pd.read_csv('InsurancePurchase.csv', sep=',', header=0)
insurance['intercept']=1
y=insurance['Purchase'].astype('category').cat.codes
X = insurance[['MOPLHOOG','MBERHOOG','MINK7512','MINK123M']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1000, random_state=1001)


# fit model 
logit_model = sm.Logit(y_test, X_test).fit()
print(logit_model.summary())

pred_prob = logit_model.predict(X)
X.loc[:,'prediction']=0

pred_prob = logit_model.predict(X_test)
X_test.loc[:,'prediction']=0
X_test.loc[pred_prob > 0.50,'prediction']=1
pd.crosstab(y_test,X_test['prediction'],rownames =['actual'],colnames=['predicted'])
```
