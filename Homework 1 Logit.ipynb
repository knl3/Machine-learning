```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
stock = pd.read_csv("StockMarket.csv", sep=',', header=0)
stock
```

```python
##Question 1-1

stock.describe(include='all')
```

```python
## Question 1-2

stock.corr(method ='pearson') 
```

```python
## Question 2-1
 
X=stock[['Year','Lag1','Lag2','Lag3','Lag4','Lag5','Volume']]
Y=stock[['Today']]
X = sm.add_constant(X)

model = sm.OLS(Y, X).fit()
predictions = model.predict(X) 
print_model = model.summary()

print_model
```

```python
## Question 2-2

X=stock[['Year','Lag1','Lag2','Lag3','Lag4','Lag5','Volume']]
X = sm.add_constant(X)

Y = stock['Direction'].astype('category').cat.codes

model2 = sm.OLS(Y, X).fit()
predictions2 = model2.predict(X) 
print_model2 = model2.summary()

print_model2
```

```python
## Question 2-3

X=stock[['Lag1','Lag2']]
X = sm.add_constant(X)
Y = stock['Direction'].astype('category').cat.codes

model3 = sm.OLS(Y, X).fit()
predictions3 = model3.predict(X) 
print_model3 = model3.summary()

print_model3
```

```python
## Question 3-1

stock['intercept'] = 1.0

X=stock[['Year','Lag1','Lag2','Lag3','Lag4','Lag5','Volume', 'intercept']]
y = stock['Direction'].astype('category').cat.codes

logit_model1 = sm.Logit(y, X).fit()
logit_model1.summary()
```

```python
## Question 3-2

stock['intercept'] = 1.0
y = stock['Direction'].astype('category').cat.codes


X=stock[['Lag1','Lag2','intercept']]
logit_model2 = sm.Logit(y, X).fit()
logit_model2.summary()
```

```python
## Question 3-3

X=stock[['Lag1','Lag2','intercept']]
y = stock['Direction'].astype('category').cat.codes
logit_model3 = sm.Logit(y, X).fit()
#print(logit_model3.summary())

pred_prob = logit_model3.predict(X)
X.loc[:,'prediction']=0
X.loc[pred_prob > 0.5,'prediction']=1
pd.crosstab(y,X['prediction'],rownames =['actual'],colnames=['predicted'])

#logit_model3.pred_table(0.5)
```

```python
## Question 3-4

stock['intercept'] = 1.0
y=stock['Direction'].astype('category').cat.codes
X=stock[['Lag1','Lag2','intercept']]

X_train = X[stock['Year'] < 2005]
X_test = X[stock['Year'] == 2005]
Y_train = y[stock['Year'] < 2005]
Y_test = y[stock['Year'] == 2005]


logit_model4 = sm.Logit(Y_train, X_train).fit()
#print(logit_model4.summary())

pred_prob = logit_model4.predict(X_test)
X_test.loc[:,'prediction']=0
X_test.loc[pred_prob > 0.50,'prediction']=1
pd.crosstab(Y_test,X_test['prediction'],rownames =['actual'],colnames=['predicted'])
```
